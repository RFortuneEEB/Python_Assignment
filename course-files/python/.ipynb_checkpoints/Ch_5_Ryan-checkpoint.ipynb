# Visualizing Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
surveys_complete = pd.read_csv('surveys_complete.csv', index_col=0)
```

```python
surveys_complete
```

```python
sns.lmplot("weight", "hindfoot_length", data=surveys_complete, fit_reg=False)
```

### sns.lmplot("x","y",data=dataframe,regression line (y/n))

```python
sns.set(font_scale=1.5)
```

### sets plot parameter for future lines

```python
sns.lmplot("weight", "hindfoot_length", data=surveys_complete, fit_reg=False)
```

```python
sns.lmplot("weight", "hindfoot_length", data=surveys_complete, fit_reg=False, height=8, aspect=1.5)
```

```python
sns.lmplot("weight", "hindfoot_length", data=surveys_complete, fit_reg=False, height=8, aspect=1.5, scatter_kws={"s":5})
```

```python
sns.lmplot("weight", "hindfoot_length", data=surveys_complete, fit_reg=False, height=8, aspect=1.5, scatter_kws={"s":5}, markers='x')
```

```python
sns.lmplot("weight", "hindfoot_length", data=surveys_complete, fit_reg=False, height=8, aspect=1.5, scatter_kws={'alpha':0.3})
```

```python
sns.lmplot("weight", "hindfoot_length", data=surveys_complete, fit_reg=False, height=8,
           aspect=1.5, scatter_kws={'alpha':0.3,"s": 50}, hue='species_id', markers='D')
```

```python
sns.lmplot("weight", "hindfoot_length", data=surveys_complete[surveys_complete.plot_id == 12], fit_reg=False, height=8,
           aspect=1.5, scatter_kws={'alpha':0.3,"s": 100}, hue='sex')
```

```python
my_fig = sns.lmplot("weight", "hindfoot_length", data=surveys_complete, fit_reg=False, height=8, 
                    aspect=1.5, scatter_kws={'alpha':0.3,"s": 50}, hue='species_id', markers='D')
my_fig.set_axis_labels('Weight (g)', 'Hindfoot Length (mm)')
```

```python
plot_dims = (14, 9)
fig, ax = plt.subplots(figsize=plot_dims)
sns.boxplot('species_id','hindfoot_length', data=surveys_complete)
ax.set(xlabel='Species ID', ylabel='Hindfoot Length (mm)')
```

```python
fig, ax = plt.subplots(figsize=plot_dims)
g = sns.violinplot('weight','species_id', data=surveys_complete, linewidth=0.2, orient="h", cut=0)
g.set_xscale('log', basex=10)
ax.set(ylabel='Species ID', xlabel='Weight (g)')
```

```python
fig, ax = plt.subplots(figsize=plot_dims)
g = sns.violinplot('weight','sex', data=surveys_complete[surveys_complete.species_id == "OL"], linewidth=0.2, orient="h", cut=0)
g.set_xscale('log', basex=10)
ax.set(ylabel='Species ID', xlabel='Weight (g)')
```

```python
fig, ax = plt.subplots(figsize=plot_dims)
sns.distplot(surveys_complete['weight'], color='c')
```

```python
fig, ax = plt.subplots(figsize=plot_dims)
sns.distplot(surveys_complete['weight'], kde=False, color='c', hist_kws=dict(edgecolor="k", linewidth=1), bins=100)
```

```python
fig, ax = plt.subplots(figsize=plot_dims)
sns.distplot(surveys_complete['weight'], kde=False, color='c', hist_kws=dict(edgecolor="k", linewidth=1), bins=40)
```

```python
from bokeh.plotting import figure 
from bokeh.io import output_notebook, show
```

```python
output_notebook()
```

```python
import numpy as np
```

```python
hist, edges = np.histogram(surveys_complete['weight'], density=True, bins=100)
my_fig = figure(title="Weight (g)",background_fill_color="#EBC8EB")
my_fig.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],fill_color="#036564", line_color="#033649")
show(my_fig)
```

```python
hist, edges = np.histogram(surveys_complete['weight'], density=True, bins=100)
my_fig = figure(title="Weight (g)")
my_fig.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],fill_color="#036564", line_color="#033649")
show(my_fig)
```

```python
from bokeh.resources import CDN
from bokeh.embed import file_html
```

```python
html = file_html(my_fig, CDN, "Weight Histogram")
out = open('weight_hist_bokeh.html','w')
out.write(html)
out.close()
```
